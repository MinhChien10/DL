```python
import numpy as np

def daoham(x):
    return 2*x + 2

def hamfx(x):
    return x*x + 2*x + 5

x = 100
numof = 1000
cost = np.zeros((numof,1))
learning_rate = 0.01
for i in range(1, numof):
    #tính giá trị dự đoán
    cost[i] = hamfx(x)
    print (x, cost[i])
    
    #gradient descent
    x -= learning_rate * daoham(x)
```
