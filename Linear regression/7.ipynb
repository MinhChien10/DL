```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/tranc/Desktop/ML_DM/DL_Tutorial-master/L1/data_square.csv').values
N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
plt.scatter(x, y)
plt.xlabel('mét vuông')
plt.ylabel('giá')

x = np.hstack((np.ones((N, 1)), x, x*x))

w = np.array([2000.,-95.,1.]).reshape(-1,1)

numOfIteration = 100000
cost = np.zeros((numOfIteration,1))
learning_rate = 0.000000002
for i in range(1, numOfIteration):
  r = np.dot(x, w) - y
  cost[i] = 0.5 * np.sum(r*r)
  w[0] -= learning_rate * np.sum(r)

  #correct the shape dimension
  w[1] -= learning_rate * np.sum(np.multiply(r, x[:,1].reshape(-1,1)))
  w[2] -= learning_rate * np.sum(np.multiply(r, x[:,2].reshape(-1,1)))
  # print(cost[i])
square = x[:,1]
predict = np.dot(x, w)
plt.plot(square,predict, 'r')
plt.show()

x1 = 110
y1 = w[0] + w[1] * x1 + w[2] * x1 * x1
print('Giá nhà cho 110m^2 là: ', y1)
```
